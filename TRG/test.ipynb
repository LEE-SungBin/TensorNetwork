```python
import numpy as np
import numpy.typing as npt
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def Bessel(n: npt.NDArray[np.int64], beta: float):
    return sp.special.iv(n, beta)

print(Bessel(-10,2))
```

```python
beta = 1/0.2
order = np.arange(30)

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(order, Bessel(order, beta) / Bessel(0, beta))
ax.set_yscale("log")
ax.set_xlim([0,None])
ax.set_ylim([1e-15,5])

plt.show()
```
